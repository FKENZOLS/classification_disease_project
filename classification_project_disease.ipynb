### Comparing and validating classifiers

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, StackingClassifier, BaggingClassifier, GradientBoostingClassifier, ExtraTreesClassifier, AdaBoostClassifier, HistGradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import roc_auc_score

classifiers = [
    RandomForestClassifier(random_state=0),
    LogisticRegression(random_state=0),
    SVC(probability=True, random_state=0),
    KNeighborsClassifier(),
    GaussianNB(),
    MLPClassifier(random_state=0),
    AdaBoostClassifier(random_state=0),
    DecisionTreeClassifier(random_state=0),
    GradientBoostingClassifier(random_state=0),
    ExtraTreesClassifier(random_state=0),
    BaggingClassifier(random_state=0),
    VotingClassifier(estimators=[('rf', RandomForestClassifier(random_state=0)), ('svc', SVC(probability=True, random_state=0)), ('lr', LogisticRegression(random_state=0))], voting='soft'),
    StackingClassifier(estimators=[('rf', RandomForestClassifier(random_state=0)), ('svc', SVC(probability=True, random_state=0)), ('lr', LogisticRegression(random_state=0))], final_estimator=RandomForestClassifier(random_state=0)),
    HistGradientBoostingClassifier(random_state=0),
]

results = []
outer_cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)

for clf in classifiers:
    inner_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

    # Initialize lists to store inner loop results
    inner_results = []
    inner_roc_aucs = []

    # Perform inner loop cross-validation
    for train_index, test_index in inner_cv.split(x, y):
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # Fit the classifier on the inner training data
        clf.fit(x_train, y_train)

        # Evaluate the classifier on the inner test data to get accuracy and ROC AUC
        accuracy_inner = clf.score(x_test, y_test)
        probas = clf.predict_proba(x_test)
        roc_auc_inner = roc_auc_score(y_test, probas[:, 1])

        inner_results.append(accuracy_inner)
        inner_roc_aucs.append(roc_auc_inner)

    # Calculate the mean accuracy and ROC AUC across all inner folds
    mean_accuracy = sum(inner_results) / len(inner_results)
    mean_roc_auc = sum(inner_roc_aucs) / len(inner_roc_aucs)

    # Evaluate the classifier using the outer loop cross-validation
    outer_results = cross_val_score(clf, x, y, cv=outer_cv)
    media = outer_results.mean() * 100

    results.append((str(clf), media, mean_roc_auc))

df = pd.DataFrame(results, columns=["Classifier", "Accuracy", "ROC AUC"])
print(df)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score
from skopt import BayesSearchCV
from sklearn.metrics import roc_auc_score

espaco_de_parametros = {
    "C": (1, 10.0),            # Regularization parameter
    "kernel": ["linear", "rbf"],  # Kernel type
    "gamma": (1e-6, 1.0, "log-uniform")  # Kernel coefficient
}

outer_cv = StratifiedKFold(n_splits=20, shuffle=True, random_state=0)  # Outer cross-validation
results = []

for train_index, test_index in outer_cv.split(x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    SVC_model = SVC(probability=True, random_state=0)

    busca = BayesSearchCV(
        SVC_model,
        espaco_de_parametros,
        n_iter=20,
        cv=3,  # Inner cross-validation with 3 folds
        random_state=0
    )
    busca.fit(x_train, y_train)

    best_model = busca.best_estimator_

    # Inner cross-validation to get accuracy and roc_auc
    inner_cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
    inner_results = cross_val_score(best_model, x_train, y_train, cv=inner_cv)

    probas = best_model.fit(x_train, y_train).predict_proba(x_train)
    roc_auc = roc_auc_score(y_train, probas[:, 1])

    media = inner_results.mean() * 100

    results.append((str(best_model), media, roc_auc))

df = pd.DataFrame(results, columns=["Classifier", "Accuracy", "ROC AUC"])
df
```
