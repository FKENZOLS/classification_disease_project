## Reduçao com KBest

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

x = data.drop(columns='diagnostico')
y = data['diagnostico']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)

skb = SelectKBest(chi2, k = 5)
skb.fit(x_train,y_train)
x_train = skb.transform(x_train)
x_test = skb.transform(x_test)


ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

# Perfomance baseline

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier()
dummy_stratified.fit(x_train, y_train)
scoredummy = dummy_stratified.score(x_test, y_test) * 100
print("A acurácia com o modelo dummy classifier é: {:.4f}%".format(scoredummy))
```

## Testando classificadores

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=0)
rfc.fit(x_train,y_train)
scorerfc = rfc.score(x_test,y_test)*100
print("A acurácia com o modelo RandomForestClassifier é: {:.4f}%".format(scorerfc))
```

```python
from sklearn.svm import SVC

svc = SVC(random_state=0)
svc.fit(x_train,y_train)
scoresvc = svc.score(x_test,y_test)*100
print("A acurácia com o modelo SVC é: {:.4f}%".format(scoresvc))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier()
knc.fit(x_train,y_train)
scoreknc = knc.score(x_test,y_test)*100
print("A acurácia com o modelo KNeighborsClassifier é: {:.4f}%".format(scoreknc))
```

## Comparando classificadores com cross validation:

```python
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

classificadores = [
    KNeighborsClassifier(),
    SVC(random_state=0),
    DecisionTreeClassifier(random_state=0),
    RandomForestClassifier(random_state=0),
    GaussianNB()

]

for clf in  (classificadores):

        cv = StratifiedKFold(n_splits = 10, shuffle=True)
        resultados = cross_val_score(clf, x, y, cv = cv)
        media = resultados.mean()*100
        print("Classificador: {}".format(clf))
        print(media)
```

O algoritmo que apresentou maior acurácia é o RandomForestClassifier.

```python
rfc=RandomForestClassifier(random_state=0)
cv = StratifiedKFold(n_splits = 10, shuffle=True)
resultados = cross_val_score(rfc, x, y, cv = cv)
media = resultados.mean()*100
print("A acurácia com o modelo RandomForestClassifier é: {}%".format(media))
```

## Melhorando as escolhas de hiperparâmetros

```python
def rfcparam(max_depth):

  modelo = RandomForestClassifier(max_depth=max_depth, random_state=0)
  modelo.fit(x_train,y_train)
  test_score = modelo.score(x_test,y_test)

  print("Arvore max_depth = {},  teste = {}".format(max_depth, test_score) )
  tabela = [max_depth, test_score]
  return tabela

resultados = [rfcparam(i) for i in range(1,50)]
resultados = pd.DataFrame(resultados, columns= ["max_depth","test"])
resultados.head()
```

```python
sns.lineplot(x = "max_depth", y = "test", data = resultados)
```

Com max_depth = 5 obtem-se o melhor resultado com o menor nivel de profundidade (depth).





```python
rfc2=RandomForestClassifier(random_state=0, max_depth=5 )
cv = StratifiedKFold(n_splits = 5, shuffle=True, random_state=0)
resultados = cross_val_score(rfc2, x, y, cv = cv)
media = resultados.mean()*100
print("A acurácia com o modelo RandomForestClassifier é: {}%".format(media))
```

Escolhendo o melhor parâmetro para o StratifiedKFold

```python
def skfparam(n_splits):

  cv = StratifiedKFold(n_splits = n_splits, shuffle=True, random_state=0)
  resultados = cross_val_score(rfc2, x, y, cv = cv)
  media = resultados.mean()*100
  print("n_split= {},  média = {}".format(n_splits, media) )
  tabela = [n_splits, media]
  return tabela

resultados2 = [skfparam(i) for i in range(2,20)]
resultados2 = pd.DataFrame(resultados2, columns= ["n_splits","media"])
```

```python
sns.lineplot(x = "n_splits", y = "media", data = resultados2)
```

O melhor numero de n_splits dentro do intervalo [2,20] é 3.




```python
#Algoritmo de maior acurácia

model = RandomForestClassifier(random_state=0, max_depth=5 )
cv = StratifiedKFold(n_splits = 3, shuffle=True, random_state=0)
resultados = cross_val_score (model, x, y, cv = cv)
media = resultados.mean()*100
print("Acurácia: {:.2f}%".format(media))
```

### Matriz de confusão:

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

rfc2.fit(x_train,y_train)
y_pred = rfc2.predict(x_test)
cm = confusion_matrix(y_test, y_pred, labels= rfc2.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rfc2.classes_)
disp.plot()


plt.show()
```
